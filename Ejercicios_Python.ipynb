```python
numero = int(input("digite un numero entero"))
divisor = 2
if numero ==2:
    print("El numero es primo")
else:
    while divisor < numero and numero % divisor!=0:
      if numero % divisor==0:
        print("El numero es primo")
      else:
        divisor = divisor +1
    if numero % (divisor-1)!=0:
      print("el número es primo")
    else:
      print("el número no es primo")


```

```python
palabra = input("Escribe una palabra")
lista = []
lista_inv=[]
for i in palabra:
  lista.append(i)
for i in range(len(lista),0,-1):
  lista_inv.append(lista[i-1])
print(lista)
print(lista_inv)

if lista == lista_inv:
  print("Es un palíndromo")
else:
  print("No es un palíndromo")

```

3. Escribir una función que reciba un número entero positivo y devuelva su factorial.

```python
def factorial(numero):
  fact = 1
  for i in range(int(numero),0,-1):
    fact = fact*i
    x= fact
  return x



factorial(5)
```

7. Escribir una función que reciba una muestra de números en una lista y devuelva otra lista con sus cuadrados.

8. Escribir una función que reciba una muestra de números en una lista y devuelva un diccionario con su media, varianza y desviación típica.

```python
def muestra(lista):
  x=0
  sum=0
  for i in range(len(lista)):
    x=x+lista[i]
  media = x/len(lista)
  for j in range(len(lista)):
    sum = (lista[j]-media)**2 + sum
  var = sum/(len(lista)-1)
  y = {"Media": media,
        "Varianza": var,
        "Desviación típica": var**(1/2)}
  return y


muestra([1,2,3,4])




```
