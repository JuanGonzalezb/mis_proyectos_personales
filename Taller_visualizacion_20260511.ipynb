##**IMPORTAR LIBRERIAS**

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

### **DATOS DE EMPPRESA**

```python
datos = {
    "Mes": ["Enero", "Febrero", "Marzo", "Abril", "Mayo"],
    "Ventas": [1200, 1500, 1100, 1800, 2000]
}

df = pd.DataFrame(datos)

print(df)
```

###GRAFICO DE BARRAS

```python
plt.bar(datos["Mes"],datos["Ventas"])

plt.title("Ventas Por Mes")
plt.xlabel("Mes")
plt.ylabel("Ventas")    
plt.show()
```

### CAMBIAR COLOR BARRAS

```python
plt.bar(datos["Mes"],datos["Ventas"], color="purple")

plt.title("Ventas Por Mes")
plt.xlabel("Mes")
plt.ylabel("Ventas")    
plt.show()
```

###**FUNCION PARA VER LOS COLORES QUE PODEMOS USAR**

```python
print(plt.colormaps())
```

###**CAMBIAR TAMAÑO**

```python
plt.figure(figsize=(4,4))
plt.bar(datos["Mes"],datos["Ventas"])
plt.title("Ventas Por Mes")
plt.xlabel("Mes")
plt.ylabel("Ventas")    
plt.show()
```
