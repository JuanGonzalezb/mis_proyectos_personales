    * Un gráfico de barras con Seaborn
3. Agregar:
    * Título
    * Color
4. Mostrar ambos gráficos

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
datos = {
    "Producto": ["Mouse", "Teclado", "Monitor", "Audifonos"],
    "Ventas": [30, 20, 15, 25]
}
```

```python
## con matplotlib

plt.bar(datos["Producto"],datos["Ventas"], color="yellow")
plt.title("Ventas Por Mes")
```

```python
## CON SEABORN

sns.barplot(x=datos["Producto"], y=datos["Ventas"], palette='deep')
plt.title("Ventas Por Mes")
```
